## 1: Perimeter
The perimeter of a rectangle is measured as 2 * width + 2 * height. Write a function that takes two arguments, width and height, and returns the perimeter of the rectangle.

```
def perimeter(width, height):
  return 2*width + 2*height

perimeter(8, 10)
```

## 2: Year 100
Write a function that asks the user to enter their name, then their age. Return a message addressed to them that tells them the year that they will turn 100 years old.

```
def year_100():
  try:
    from datetime import datetime
  except ImportError:
        raise NotImplementedError("You need the datetime package to figure out when you will turn 100")
  else:
    name = input('Hi There, What is your name?')
    print('Hi, {}'.format(name))
    age = int(input('What is your age, {}?'.format(name)))
    date_100 = datetime.now().year + 100 - age
    print('Did you know you will turn 100 in the year {}'.format(date_100))

year_100()
```
