```
from time import time
from time import sleep
from random import randint
from IPython.core.display import clear_output
from warnings import warn
from bs4 import BeautifulSoup
import requests

def process_page(soup, news): 
  
  raw_news = soup.select('.search-result-news')

  for new in raw_news:
    headline = new.select_one('.content > header > h3').get_text().strip()
    author = new.select_one('.content > header > p > span > span').get_text().strip()
    synopsis = new.select_one('.content > p').get_text().strip()
    date_time = str(new.select_one('.content > header > p > time'))[64:83].strip()
    n = {'headline': headline, 'author': author, 'synopsis': synopsis, 'date_time' : date_time}
    news.append(n)
    

start_time = time()
request_count = 0

news = []

has_next_page = True
MAX_REQUESTS = 3 
page_number = 1
query = {'news': page_number}
url = 'https://www.space.com/news'
headers = {'user-agent': 'newsscraper - school project'}

while has_next_page and request_count <= MAX_REQUESTS:
 
  clear_output(wait = True)
  
  response = requests.get(url, params=query, headers=headers)

  if(response.ok):
    
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    process_page(soup, news)

    next_button = soup.select('.listings-pagination-button')
    has_next_page = len(next_button) > 0
    
  else:
    warn('Request #: {}, Failed with status code: {}'.format(request_count, response.status_code))
  
  request_count += 1
  
  sleep(randint(1,3))
  
  elapsed_time = time() - start_time
  print('Requests: {}, Frequency: {} requests/s, {} stories processed.'.format(request_count, request_count/elapsed_time, len(news)))
  
  page_number += 1
      
print('Sraping complete')
print('Requests: {}, Frequency: {} requests/s, {} stories processed.'.format(request_count, request_count/elapsed_time, len(news)))
print(news[0:5])
```
